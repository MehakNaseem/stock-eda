```python
# Importing all the libraries we need
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Downloading stock data for Apple, Tesla and Google for the last 2 years
stocks = ['AAPL', 'TSLA', 'GOOGL']
data = yf.download(stocks, period='2y')
print(data.head())
```

```python
# Let's see the shape of our data (rows and columns)
print("Shape of data:", data.shape)

# Check for any missing values
print("\nMissing values:\n", data.isnull().sum())

# Basic statistics
print("\nBasic Statistics:\n", data.describe())
```

```python
# Plotting the closing price of all 3 stocks over time
data['Close'].plot(figsize=(12, 6))
plt.title('Stock Closing Prices Over Last 2 Years')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend(['AAPL', 'TSLA', 'GOOGL'])
plt.grid(True)
plt.show()
```

```python
# Calculate 50-day and 200-day moving averages for Apple
aapl = data['Close']['AAPL']

aapl_ma50 = aapl.rolling(window=50).mean()
aapl_ma200 = aapl.rolling(window=200).mean()

# Plot them together
plt.figure(figsize=(12, 6))
plt.plot(aapl, label='AAPL Price')
plt.plot(aapl_ma50, label='50-Day MA')
plt.plot(aapl_ma200, label='200-Day MA')
plt.title('Apple Stock - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate daily returns for all 3 stocks
daily_returns = data['Close'].pct_change()

# Plot them
daily_returns.plot(figsize=(12, 6))
plt.title('Daily Returns of Stocks')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend(['AAPL', 'TSLA', 'GOOGL'])
plt.grid(True)
plt.show()
```

```python
# Plot trading volume for all 3 stocks
data['Volume'].plot(figsize=(12, 6))
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend(['AAPL', 'TSLA', 'GOOGL'])
plt.grid(True)
plt.show()
```

```python
# See how closely the 3 stocks move together
correlation = data['Close'].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Between Stocks')
plt.show()
```
